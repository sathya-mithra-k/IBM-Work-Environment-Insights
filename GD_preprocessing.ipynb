```python
df1 = df.drop(['Unnamed: 12','Unnamed: 13','Unnamed: 14','Unnamed: 15','Unnamed: 16','Unnamed: 17','Unnamed: 18','Unnamed: 19','Unnamed: 20','Unnamed: 21'],axis = 1)
```

```python
df1[['Employment Tenure','Location']] = df1[['Employment Tenure','Location']].fillna("Unknown")
df1.head()
```

```python
df1['Date'] = pd.to_datetime(df1['Date'], errors='coerce')
```

```python
df1['Recommend'] = df1['Recommend'].map({'Yes': 1, 'No': 0})
df1['CEO Approval'] = df1['CEO Approval'].map({'Yes': 1, 'No': 0})
df1['Business Outlook'] = df1['Business Outlook'].map({'Yes': 1, 'No': 0})
```

```python
df1.head()
```

```python
from textblob import TextBlob
```

```python
def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

df1['Pros_Sentiment'] = df1['Pros'].apply(get_sentiment)
df1['Cons_Sentiment'] = df1['Cons'].apply(get_sentiment)
```
